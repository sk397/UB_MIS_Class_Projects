```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt

# load data
housing = pd.read_csv('housing.csv')
```

```python
# show data type for the data set
housing.dtypes
```

```python
# change WATERFRONT data type to categorical
housing['waterfront'] = pd.Categorical(housing.waterfront)
```

```python
# check the data type has changed
print(housing.waterfront.dtype)
print(housing.waterfront.cat.categories)
```

```python
# change ID and zipcode to data type to string or categorical data
housing.ID = housing.id.astype('string')
housing['zipcode'] = housing['zipcode'].astype('category')
```

```python
# create a subset of data with only numerical variables
pcahousing = housing[[ 'price', 'bedrooms', 'bathrooms', 'condition',
                       'sqft_living', 'sqft_lot', 'floors', 'grade', 
                       'sqft_above','sqft_basement', 'sqft_living15', 'sqft_lot15', 'view']]
```

```python
#Import to do PCA
from sklearn.decomposition import PCA
from sklearn import preprocessing


# create principal components
pcs = PCA()
pcs.fit(pcahousing)
```

```python
# create principal components
pcs = PCA()
pcs.fit(pcahousing)
pcsSummary_df = pd.DataFrame({'Standard deviation':
np.sqrt(pcs.explained_variance_),
'Proportion of variance':
pcs.explained_variance_ratio_,
'Cumulative proportion':
np.cumsum(pcs.explained_variance_ratio_)})
pcsSummary_df = pcsSummary_df.transpose()
pcsSummary_df.columns = ['PC{}'.format(i) for i in range(1,
len(pcsSummary_df.columns) + 1)]
pcsSummary_df.round(4)
```

```python
pcsComponents_df = pd.DataFrame(pcs.components_.transpose(),
columns=pcsSummary_df.columns,
index=pcahousing.columns)
pcsComponents_df
```

```python
# use the preprocessing function to normalize the data
pcs.fit(preprocessing.scale(pcahousing))
```
